```r
# install packages
suppressPackageStartupMessages(library(tidyverse))
suppressPackageStartupMessages(library(magrittr))
suppressPackageStartupMessages(library(mosaic))
suppressPackageStartupMessages(library(tigerstats))
# get data set
cabrera <- tigerstats::cabrera
```

The `cabrera` data set has hits for MLB All-Star Miguel Cabrera from 2009-2012. Read more about it [here](https://www.rdocumentation.org/packages/tigerstats/versions/0.2.7/topics/cabrera)

```r
# view data set
head(cabrera, 10)
```

## Create Spray Chart

```r
# turn off warnings for chart ----- 
options(warn=-1)
# plot spray chart ----- 
cabrera %>% 
  ggplot2::ggplot(aes(x = hitx, y = hity)) + 
  ggplot2::geom_point(aes(color = hit_outcome)) + 
  ggplot2::coord_equal() + 
  ggplot2::facet_wrap(~ season) +
  ggplot2::labs(
    title = "Spray Chart of Miguel Cabreras balls hit into play by season (2009-2012",
    subtitle = "from tigerstats package")
# turn warnings back on -----
options(warn=-0)
```
